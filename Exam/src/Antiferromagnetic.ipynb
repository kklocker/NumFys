```python
import matplotlib.pyplot as plt
import palettable as pl
emr = pl.cartocolors.sequential.Emrld_7_r.get_mpl_colormap()
plt.style.use('Solarize_Light2')
```

```python
from mon_jasnow import metropolis_subroutine
from lattice_utils import basic_lattice
```

```python
J = -1.0
N = 200
T=0.1
sweeps = 100
```

```python
lattice = basic_lattice(N, False)
init_lattice = lattice.copy()
for i in range(sweeps):
    metropolis_subroutine(lattice, T, bc='torus', J=J)

fig, ax = plt.subplots(nrows=2, ncols=1, tight_layout=True, figsize=(10, 10))
ax[0].imshow(init_lattice, cmap=emr)
ax[0].grid()
ax[0].set_axis_off()
ax[1].imshow(lattice, cmap=emr)
ax[1].grid()
ax[1].set_axis_off()

fig.savefig("../img/antiferro.pdf", bbox_inches = 'tight',
    pad_inches = 0)
```
