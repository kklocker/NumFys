```python
U_dist_linear = np.linspace(0,1, 2000)
```

```python
U_dist_1 = Ur(last_step1, particle_1['alpha'])
U_dist_2 = Ur(last_step2, particle_2['alpha'])
```

```python
fig = plt.figure(figsize=(5, 5), tight_layout=True)
a1 = plt.hist(U_dist_1, density=True, bins = 100, label= fr"$\Delta U = 10k_BT$")
a2 = plt.hist(U_dist_2, density=True, bins = 100,label= fr"$\Delta U = 0.1k_BT$")
# plt.close()
plt.plot(a1[1][1:], a1[0])
plt.plot(a2[1][1:], a2[0])
plt.legend(fontsize=15)
plt.xticks(size =15)
plt.yticks(size =15)
plt.xlabel(fr"$\hat U$", size=20)
plt.ylabel(fr"$\rho(\hat U )$", size=20)
plt.savefig(f"img/hist_N._{N}.png", facecolor=fig.get_facecolor())
```

```python
p1 = boltzmann(U_dist_linear, particle_1['delta_u'], particle_1['kbt']) * particle_1['kbt']
p2 = boltzmann(U_dist_linear, particle_2['delta_u'], particle_2['kbt']) * particle_2['kbt'] 
```

```python
rho_u1 = normalize_distribution(U_dist_linear * p1)
rho_u2 = normalize_distribution(U_dist_linear * p2)
```

```python
ux1 = np.linspace(0, 1, a1[0].shape[0])
rr1 = normalize_distribution(ux1 * a1[0])
ux2 = np.linspace(0, 1, a2[0].shape[0])
rr2 = normalize_distribution(ux2 * a2[0])
```

```python
fig = plt.figure(figsize=(5, 5), tight_layout=True)
plt.plot(ux2, rr1, label = fr"$\Delta U  = 10k_BT$")
plt.plot(U_dist_linear, rho_u1, label = "Boltzmann ")
plt.legend(fontsize = 15)
plt.xlabel(fr"$\hat U$", size = 20)
plt.ylabel(fr"$\hat U\rho(\hat U)$", size = 20)
plt.xticks(size =15)
plt.yticks(size =15)
plt.savefig(f"dist_1_N_{N}.png", facecolor=fig.get_facecolor())
```

```python
fig = plt.figure(figsize=(5, 5), tight_layout=True)
plt.plot(ux2, rr2, label =fr"$\Delta U  = 0.1k_BT$")
plt.plot(U_dist_linear, rho_u2,  label = "Boltzmann")
plt.legend(fontsize = 15)
plt.xlabel(fr"$\hat U$", size = 20)
plt.ylabel(fr"$\hat U \rho(\hat U)$", size = 20)
plt.xticks(size =15)
plt.yticks(size =15)
plt.savefig(f"dist_2_N_{N}.png", facecolor=fig.get_facecolor())
```
