```python
from eqsolver import get_matrix
from grid import get_grid,normalize, convert_to_dict
from classification import contains
import numpy as np
import os
```

```python
%%time
g, p = 4, 3
pos_list = get_grid(g, p)
m_ = normalize(pos_list)
print(m_)
p_dict = convert_to_dict(pos_list)
c_path = f"solutions/classifications/{g}_{p}"
if os.path.exists(c_path + ".npy"):
    classification = np.load(c_path + ".npy")
else:
    classification = contains(p_dict, m_)
    np.save(c_path, classification)
e,v = np.load(f"solutions/solution_{g}_{p}.npy", allow_pickle=True)
```

```python
e_ = np.sign(e[0]) * e
```

```python
e0 = np.abs(e_[0])
args = np.argsort(np.sign(e0)*e_/e0)
tmp = np.sort(np.sign(e0)*e_/e0)
E = tmp[tmp>=0.99999] * e0
```

```python
omega_0 = np.sqrt(e0)/2.100
results = (np.sqrt(E)/omega_0).astype(float)
```

```python
V = np.sign(e0)*(v[:,args])[:,tmp>=0.999]
```

```python
N_plot = min(V.shape[1], 6)
cases = V[:,:N_plot]

plot_list = [*pos_list, pos_list[0]]


E = e[0]

def trim_axs(axs, N): 
    """
    Reduce *axs* to *N* Axes. All further Axes are removed from the figure.
    Stolen from https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/markevery_demo.html#sphx-glr-gallery-lines-bars-and-markers-markevery-demo-py
    """
    axs = axs.flat
    for ax in axs[N:]:
        ax.remove()
    return axs[:N]
v_max= float(np.max(V[:, 0]))
v_min = -v_max
ncols = 2
nrows = N_plot // ncols + 1
print(nrows, ncols)
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, sharex=True, sharey=True, figsize=(15, 10))
axs = trim_axs(axs, N_plot)

for i in range(N_plot):
    w = np.ma.array(V[:,i].reshape((m_,m_)),mask = np.invert(classification))
    axs[i].contourf(w.T / E, cmap = emr, levels=100, corner_mask=False)
    #axs[i].legend([r"$\frac{\epsilon_i}{\epsilon_0}=$" + f"{np.real(e[i]/E):.4f}"], fontsize = 15,loc = "best")
    axs[i].legend([f"{i}"], fontsize = 15)
    axs[i].plot(*zip(*plot_list), 'k', lw=1)
    axs[i].set_yticklabels([])
    axs[i].set_ylim(-1,m_+1)
    axs[i].set_xlim(-1,m_+1)
    axs[i].set(aspect='equal')
    
    #plt.colorbar(im, ax=axs[i])
    #axs[i].set_title()
fig.subplots_adjust(wspace=0.5, hspace=0)
fig.tight_layout()
#fig.colorbar(ax=axs)
#ax[i,j].plot([1, 2])
```
